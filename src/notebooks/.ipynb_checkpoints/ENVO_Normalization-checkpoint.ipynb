```python
import pandas as pd
#import vaex
import os
from runner import runner
```

```python
outputDir = '../../nlp/data/output/'
outFile = 'broadMediumLocal.tsv'
inputDir = '../../nlp/data/input/'
targetDir = '../../target/'
inFile = 'harmonized-table.parquet.gz'
settingsFile = '../../nlp/settings.ini'
termlistFolder = '../../nlp/terms'
cols = ['id', 'accession_biosample_id','env_broad_scale','env_medium','package_name','env_local_scale']
```

```python
%%time
harmonizedDF = pd.read_parquet(os.path.join(targetDir, inFile))
harmonizedDF.head()
```

```python
subsetHarmonizedDF = harmonizedDF[cols]
subsetHarmonizedDF = subsetHarmonizedDF[~subsetHarmonizedDF['env_broad_scale'].isnull()]
```

```python
subsetHarmonizedDF.head()
```

```python
subsetNonEnvo = subsetHarmonizedDF[~subsetHarmonizedDF['env_broad_scale'].str.contains('ENVO:\d')]
subsetNonEnvo.head()
```

```python
broad_scale = list(subsetNonEnvo['env_broad_scale'].unique())
```

```python
medium = list(subsetNonEnvo['env_medium'].unique())
```

```python
local = list(subsetNonEnvo['env_local_scale'].unique())
```

```python
def list2DF(l, prefix):
    columns = ['id' , 'text']
    df = pd.DataFrame(columns=columns)
    df['id'] = (range(1,len(l)+1))
    df['id'] = prefix+'_'+df['id'].astype(str)
    df['text'] = pd.Series(l)
    return df
```

```python
broadNlpInput = list2DF(broad_scale, 'broad')
mediumNlpInput = list2DF(medium, 'medium')
localNlpInput = list2DF(local, 'local')
localNlpInput.head(20)
```

```python
nlpInput =pd.concat([broadNlpInput, mediumNlpInput, localNlpInput])
nlpInput.to_csv(os.path.join(inputDir, outFile), sep='\t', index=None)
```

```python
%%time
runner.json2tsv(input=os.path.join(inputDir, 'envo.json'),output=os.path.join(termlistFolder, 'envo'))
runner.prepare_termlist(input=os.path.join(termlistFolder, 'envo_nodes.tsv'), output=os.path.join(termlistFolder, 'envo_termlist.tsv'))
runner.run_oger(settings=settingsFile)
```

```python
ogerOutput = pd.read_csv(os.path.join(outputDir, 'runNER_Output.tsv'), sep='\t', low_memory=False)
ogerOutput = ogerOutput.drop_duplicates()
ogerOutput.head()
```

```python
termEntityId = ogerOutput[['DOCUMENT ID', 'MATCHED TERM', 'PREFERRED FORM', 'ENTITY ID']]
termEntityId = termEntityId.drop_duplicates()
termEntityId.head()
```

```python
termEntityId.to_csv(os.path.join(outputDir, 'termEntityId.tsv'), sep='\t', index=None)
```

```python
mergedDF = localNlpInput.merge(termEntityId, how='left', left_on='id', right_on='DOCUMENT ID')
mergedDF.head()
```

```python
mergedDF.to_csv(os.path.join(outputDir, 'mergedDF.tsv'), sep='\t', index=None)
```
