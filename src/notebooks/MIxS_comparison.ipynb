```python
import pandas as pd
import os
import yaml
import csv
```

```python
parquetFile = pd.read_parquet(path='../input/harmonized-table.parquet.gzip')
parquetFile.head()
```

```python
parquetFileColumns = set(parquetFile.columns)
parquetFileColumns
```

```python
#Schema folder is basically https://github.com/cmungall/mixs-source/tree/main/src/schema
schemaFolder = '../input/schema'
possibleColumnNames = []
for yamlFile in os.listdir(schemaFolder):
    data = yaml.load(open(os.path.join(schemaFolder, yamlFile), 'r'), yaml.Loader)
    term = yamlFile.strip('.yaml').replace('_','-')
    if data.get('classes').get(term) is not None:
        slots = data.get('classes').get(term).get('slots')
        possibleColumnNames.extend(slots)
        
possibleColumnNames = set(possibleColumnNames)
print(len(possibleColumnNames))
```

```python
# Column names that are MIxS

commonList = list(parquetFileColumns & possibleColumnNames)
print(len(commonList))
```

```python
# Non-MIxS terms
differenceList = possibleColumnNames.difference(parquetFileColumns)
print(len(differenceList))
```

```python
pd.DataFrame(commonList).to_csv('../output/MIxS_Columns.tsv', sep='\t', index=None, header=None)
pd.DataFrame(differenceList).to_csv('../output/Non_MIxS_Columns.tsv', sep='\t', index=None, header=None)
```
