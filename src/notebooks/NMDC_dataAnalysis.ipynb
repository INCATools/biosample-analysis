```python
import pandas as pd
```

```python
outputDir = '../../nlp/data/nlp/output/'
outFile = 'runNER_Output.tsv'
inputDir = '../../downloads/'
inFile = 'nmdc-biotable-for-ner-20201016.tsv'

runnerOutput = pd.read_csv(outputDir+outFile, sep='\t' )
runnerOutput = runnerOutput.drop(['START POSITION', 'END POSITION', 'ZONE','SENTENCE ID', 'ORIGIN', 'UMLS CUI', 'SENTENCE'], axis = 1)
runnerOutput = runnerOutput.drop_duplicates()
runnerOutput['CURIE_SYNONYM_REMOVED'] = runnerOutput['ENTITY ID'].str.strip('_SYNONYM').str.replace(':','_')
runnerOutput.head()
```

```python
ncbiInput = pd.read_csv(inputDir+inFile, sep='\t', low_memory=False, usecols=['GOLD_ID', 'BROAD_SCALE_LABEL', 'LOCAL_SCALE_LABEL', 'MEDIUM_LABEL', 'ENV_BROAD_SCALE', 'ENV_LOCAL_SCALE', 'ENV_MEDIUM'] )
ncbiInput.head()
```

```python
nlpInput = pd.read_csv('../../nlp/data/nlp/input/nlpNCBIInput.tsv', sep='\t', low_memory=False, index_col=None)
nlpInput.head()
```

```python
inputOutputJoin = ncbiInput.merge(runnerOutput, left_on='GOLD_ID', right_on='DOCUMENT ID', how='left' )
inputOutputJoin.head()
```

```python
inputOutputJoin = inputOutputJoin.drop(['DOCUMENT ID'], axis=1)
#inputOutputJoin = inputOutputJoin.drop_duplicates()
inputOutputJoin['CURIE'] = inputOutputJoin['ENTITY ID']
inputOutputJoin = inputOutputJoin.drop(['ENTITY ID'], axis=1)
inputOutputJoin.head()
```

```python
inputOutputJoin.to_csv(outputDir+'inputOutputJoined.tsv', sep='\t', index=None)
```

```python
filteredMatch = inputOutputJoin.query('CURIE_SYNONYM_REMOVED == ENV_BROAD_SCALE' and \
                                      'CURIE_SYNONYM_REMOVED == ENV_LOCAL_SCALE' and \
                                      'CURIE_SYNONYM_REMOVED == ENV_MEDIUM')

#filteredMatch = filteredMatch.drop_duplicates()
```

```python
filteredMatch.head()
```

```python
filteredMatch.to_csv(outputDir+'filteredMatch.tsv', sep='\t', index=None)
```

```python
# env_{local,medium,broad}_matched
inputOutputJoin['env_broad_matched'] = inputOutputJoin['CURIE_SYNONYM_REMOVED'] == inputOutputJoin['ENV_BROAD_SCALE']
inputOutputJoin['env_local_matched'] = inputOutputJoin['CURIE_SYNONYM_REMOVED'] == inputOutputJoin['ENV_LOCAL_SCALE']
inputOutputJoin['env_medium_matched'] = inputOutputJoin['CURIE_SYNONYM_REMOVED'] == inputOutputJoin['ENV_MEDIUM']
inputOutputJoin['false_positives'] = (inputOutputJoin.iloc[:,-3:]==False).all(True)

inputOutputJoin.head(5)
```

```python
grpList = ['GOLD_ID', 'BROAD_SCALE_LABEL', 'LOCAL_SCALE_LABEL', 'MEDIUM_LABEL','ENV_BROAD_SCALE','ENV_LOCAL_SCALE','ENV_MEDIUM', 'TYPE', 'MATCHED TERM', 'PREFERRED FORM', 'CURIE_SYNONYM_REMOVED', 'CURIE']

collapsedIO  = inputOutputJoin.groupby(grpList).any()
```
